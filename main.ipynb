```python
import json
from collections import Counter

# Đọc file train.json từ local
with open('train.json', 'r', encoding='utf-8') as f:
    train_data = [json.loads(line) for line in f]

# Lấy danh sách Toxicity từ train data
toxicity_values = [item["Toxicity"] for item in train_data]

# Đếm số lượng
Counter(toxicity_values)
```
